## Data Wrangling


### General Properties

```python
# Loading data
df=pd.read_csv('tmdb-movies.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
# Let's create a function for sorting data
def separate(df):
    return df.str.split('|', expand=True)
def ax_and_titles(y,x,t):
    return plt.ylabel(y),plt.xlabel(x),plt.title(t)
```

### Data Cleaning 

```python
# Checking for duplicates
df.duplicated().sum()
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
# Droping columns which will not be used
df.drop(['homepage','tagline','keywords','production_companies'], axis=1, inplace=True)
```

```python
df.info()
```

```python
# Checking NAs in the director column for clues
df[df['director'].isnull()]
```

```python
# We didn't find any clues for NAs, so we'll just drop them
df.dropna(inplace=True)
```

```python
df.info()
```

```python
# Fixing datatypes 
df['release_date']=pd.to_datetime(df['release_date'])
```

```python
df.head()
```

```python
#Let's see correlation matrix
pd.plotting.scatter_matrix(df, figsize=(10,10));
```

```python
# For further investigation we need to work with genres column, let's check how many unique values it has
df.genres.value_counts()
```

```python
# Let's assume that is the first mentioned genre in a row is the main one, so we can make our data a bit cleaner
df['genres']=df['genres'].apply(lambda x:x.split("|")[0])
df.genres.value_counts()
```

## Exploratory Data Analysis


### Which genres are most popular from year to year?

```python
# Maximum average popularity per genre
df.groupby(['genres'])['popularity'].mean().max()
```

<a id='link1'></a>

```python
# Average popularity per genre 
df.groupby(['genres'])['popularity'].mean().sort_values(ascending=False).plot(kind='bar',);
ax_and_titles('Popularity','Genres','Average popularity per genre');
```

```python
# Let's calculate profit/loss in a new column 
df['profit/loss']=df['revenue']-df['budget']
```

### What kinds of properties are associated with movies that have high revenues?


```python
# Let's check revenue statistics
df['revenue'].describe()
```

```python
# Let's clean revenue from zero values in a new dataframe and recheck its distribution
revenue=df[df['revenue']>0]
revenue.describe()
```

```python
# Let's see correlation matrix
pd.plotting.scatter_matrix(revenue, figsize=(12,12));
```

```python
#Let's see how quantity of votes is correlating with gained revenue
revenue.plot(x='vote_count', y='revenue', kind='scatter', figsize=(6,6));
ax_and_titles('Revenue','Vote count','Vote_count vs Revenue Correlation');
```

##### Here we can see slight correlation between vote_count and revenue, generally speaking the more tickets were bought the more votes were left. 

<a id='link2'></a>

```python
# Let's see the revenue average per genre
revenue.groupby('genres')['revenue'].mean().sort_values(ascending=False).plot(kind='bar');
ax_and_titles('Revenue','genres','Revenue Average per genre');
```

##### Here we can see genres which produce the highest revenue 

<a id='link3'></a>

```python
# Let's see average profitability per genre
df.groupby(['genres'])['profit/loss'].mean().sort_values(ascending=False).plot(kind='bar');
ax_and_titles('Profit','Genres','Average Profitability per genre');
```

##### The top 5 genres in this plot are almost in the same order as in Average revenue plot, what confirms the obvious positive correlation between higher revenue and higher profit 

```python
# Let's divede our revenue dataframe for revenue below median and above
revenue_median=revenue['revenue'].median()
low_revenue=revenue[revenue['revenue']<revenue_median]
high_revenue=revenue[revenue['revenue']>=revenue_median]
revenue['id'].count()==low_revenue.shape[0]+high_revenue.shape[0]
```

```python
#Profitability distribution comparison
plt.hist(low_revenue['profit/loss'],color='b', alpha=.5, label='Low Revenue')
plt.hist(high_revenue['profit/loss'],color='g', alpha=.5, label='High Revenue')
plt.legend();
plt.ylabel('Quantity')
plt.xlabel('Profit/Loss')
plt.title('Profitability distribution per high and low revenue movies');
```

##### Here we have more skewed to the right high revenue profit distribution, what also confirms the obvious positive correlation between higher revenue and higher profit

```python
#Runtime distribution comparison
plt.hist(low_revenue['runtime'],color='b', alpha=.5, label='Low Revenue')
plt.hist(high_revenue['runtime'],color='g', alpha=.5, label='High Revenue')
plt.legend();
plt.ylabel('Quantity')
plt.xlabel('Runtime')
plt.title('Runtime distribution per high and low revenue movies');
```

##### Here we can see that films with higher revenue had longer runtime

<a id='link4'></a>

```python
#Vote Average distribution comparison
plt.hist(low_revenue['vote_average'],color='b', alpha=.5, label='Low Revenue')
plt.hist(high_revenue['vote_average'],color='g', alpha=.5, label='High Revenue')
plt.legend();
plt.ylabel('Quantity')
plt.xlabel('Vote Average')
plt.title('Vote Average distribution per high and low revenue movies');
```

###### The voter_average pattern looks almost the same for low and high revenue movies, we can say here that movies with lower revenue level can bring viewers the same level of satisfaction as with high revenue level 

```python
#Vote count distribution comparison
plt.hist(low_revenue['vote_count'],color='b', alpha=.5, label='Low Revenue')
plt.hist(high_revenue['vote_count'],color='g', alpha=.5, label='High Revenue')
plt.legend();
plt.ylabel('Quantity')
plt.xlabel('Vote Count')
plt.title('Vote Count distribution per high and low revenue movies');
```
