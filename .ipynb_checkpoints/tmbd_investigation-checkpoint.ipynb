## Data Wrangling


### General Properties

```python
# Loading data
df=pd.read_csv('tmdb-movies.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

### Data Cleaning 

```python
# Checking for duplicates
df.duplicated().sum()
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
# Droping columns which will not be used
df.drop(['homepage','tagline','keywords','production_companies'], axis=1, inplace=True)
```

```python
df.info()
```

```python
# Checking nulls in director column for clues
df[df['director'].isnull()]
```

```python
# Drop NAs
df.dropna(inplace=True)
```

```python
df.info()
```

```python
# Fixing datatypes 
df['release_date']=pd.to_datetime(df['release_date'])
```

```python
df.head()
```

```python
df.hist(figsize=(20,20));
```

```python
# For further investigation we need to work with genres column, let's check how many unique values it has
df.genres.value_counts()
```

```python
# Let's assume that is the first mentioned genre in a row is the main one, so we can make our data a bit cleaner
df['genres']=df['genres'].apply(lambda x:x.split("|")[0])
df.genres.value_counts()
```
